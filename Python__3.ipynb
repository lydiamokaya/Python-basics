```
#loadind our CellGeo datasets
cellsGeoDescription = pd.read_excel("cells_geo_description.xlsx", index_col= None)
cellsGeoDescription
```

```
cellGeo = pd.read_csv("cells_geo.csv", sep = ';', keep_default_na=False, na_values=[""])
cellGeo.head()
```

```
cellGeo.info(verbose=None, buf=None, max_cols=None, memory_usage=None, null_counts=None)
```

```
cellGeo.rename(columns={"SITE_CODE":"SITE_ID"}, inplace= True)
cellGeo.head()
```

```
#droping duplicate rows
cellGeo1 = cellGeo.drop_duplicates(subset="VILLES", keep="first")
cellGeo1.index = range(len(cellGeo1.index))
cellGeo1.head()
```

```
#droping unneccesary columns
CellGeo1 =cellGeo1.loc[:,['VILLES','STATUS','SITE_ID']]
CellGeo1.head()
```

```
#loading our telecom datasets
CDR_description = pd.read_excel("CDR_description.xlsx", index_col= None)
CDR_description
```

```
#day one CDR
telcomDataset = pd.read_csv("Telcom_dataset.csv", index_col= None)
telcomDataset.head()
```

```
#Dropping null values in dataset 1
telcomDataset = telcomDataset.dropna()
telcomDataset.index = range(len(telcomDataset.index))
telcomDataset.head()
```

```
telcomDataset.info(verbose=None, buf=None, max_cols=None, memory_usage=None, null_counts=None)
```

```
#day two CDR
telcomDataset2 = pd.read_csv("Telcom_dataset2.csv", index_col= None)
telcomDataset2.head()
```

```
#Dropping null values in dataset 2
telcomDataset2 = telcomDataset2.dropna()
telcomDataset2.index = range(len(telcomDataset2.index))
telcomDataset2.tail()
```

```
telcomDataset2.info(verbose=None, buf=None, max_cols=None, memory_usage=None, null_counts=None)
```

```
#day three CDR
telcomDataset3 = pd.read_csv("Telcom_dataset3.csv", index_col= None)
telcomDataset3.head()
```

```
#Dropping null values in dataset 3
telcomDataset3 = telcomDataset3.dropna()
telcomDataset3.index = range(len(telcomDataset3.index))
telcomDataset3.tail()
```

```
telcomDataset2.info(verbose=None, buf=None, max_cols=None, memory_usage=None, null_counts=None)
```

```
#dropping unneccesary columns dataset1
new_telcomDataset = telcomDataset.drop(["DW_A_NUMBER_INT", "DW_B_NUMBER_INT", "COUNTRY_A", "COUNTRY_B", "CELL_ID"], axis= 1)
new_telcomDataset.head()
```

```
#renaming column names
new_telcomDataset.rename(columns={"PRODUTC":"PRODUCT", "DATETIME":"DATE_TIME"}, inplace= True)
new_telcomDataset.head()
```

```
#dropping unneccesary columns dataset2
new_telcomDataset2 = telcomDataset2.drop(["DW_A_NUMBER", "DW_B_NUMBER", "COUNTRY_A", "COUNTRY_B", "CELL_ID"], axis= 1)
new_telcomDataset2.head()
```

```
#dropping unneccesary columns dataset3
new_telcomDataset3 = telcomDataset3.drop(["DW_A_NUMBER_INT", "DW_B_NUMBER_INT", "COUNTRY_A", "COUNTRY_B", "CELLID"], axis= 1)
new_telcomDataset3.head()
```

```
#replacing column name
new_telcomDataset3.rename(columns={"SIET_ID":"SITE_ID"}, inplace= True)
new_telcomDataset3.head()
```

```
#merging the 3 datasets
merged_telcomdatasets = pd.concat([new_telcomDataset,new_telcomDataset2,new_telcomDataset3], axis=0, join= 'outer',levels=None, names=None, verify_integrity= False, copy=True)
merged_telcomdatasets.index = range(len(merged_telcomdatasets.index))
merged_telcomdatasets.head(30)
```

```
#Removing the special characters in SITE_ID
spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~",""]
for char in spec_chars:
    merged_telcomdatasets['SITE_ID'] = merged_telcomdatasets['SITE_ID'].str.replace(char, ' ')
```

```
merged_telcomdatasets.head(30)
```

```
#merging the CellGeo dataset with the merged 3 telcom datasets
MTN_final= CellGeo1.merge(merged_telcomdatasets, on='SITE_ID', how='inner')
MTN_final.head(30)
```

```
#spliting the DATE_TIME column to 2 columns
MTN_final['DATE_TIME'] = pd.to_datetime(MTN_final['DATE_TIME'])
MTN_final['DATE'], MTN_final['TIME'] = MTN_final['DATE_TIME'].dt.normalize(), MTN_final['DATE_TIME'].dt.time
print(MTN_final)
```
